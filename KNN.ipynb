```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier


from sklearn.metrics import accuracy_score


df = pd.read_table('iris.data.csv', delimiter =",") 
df["species"] = df["Iris-setosa"]
df["Slength"] = df["5.1"]
df["Swidth"] = df["3.5"]
df["Plength"] = df["1.4"]
df["Pwidth"] = df["0.2"]
df = df.drop('5.1',axis=1)
df = df.drop('3.5',axis=1)
df = df.drop('1.4',axis=1)
df = df.drop('0.2',axis=1)
df = df.drop('Iris-setosa',axis=1)
df['species']=df['species'].map({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
df.head()
```

```python
x=df[['Slength','Swidth','Plength','Pwidth']]
y=df['species']
scores =[]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30) #split our data with test size of 20% 

for i in range(1,30):
    knn=KNeighborsClassifier(n_neighbors=i) 
    knn.fit(x_train,y_train) 
    y_pred=knn.predict(x_test) 
    print(f'for {i} to Acuuracy=',accuracy_score(y_pred,y_test))
    scores.append(round(accuracy_score(y_pred,y_test),2))
```
